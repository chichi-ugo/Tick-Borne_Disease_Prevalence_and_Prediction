```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
import io
import pandas as pd

import warnings
warnings.filterwarnings('ignore')

from pathlib import Path
from collections import Counter
```

```python
data = files.upload()
```

```python
# Loading data into a dataframe
tick_pathogen_data_df = pd.read_csv(io.BytesIO(data['Ixodes_Pathogens_County_Table_2021.csv']))
tick_pathogen_data_df
```

```python
tick_pathogen_data_df.columns
```

```python
# Removing unneeded columns
tick_path_df = tick_pathogen_data_df.drop(['FIPS_Code',
                                           'County',
                                           'Borrelia_burgdorferi_sensu_stricto_Data_Source',
                                           'Borrelia_mayonii_Data_Source',
                                           'Borrelia_miyamotoi_Data_Source',
                                           'Anaplasma_phagocytophilum_Data_Source',
                                           'Ehrlichia_muris_eauclairensis_Data_Source',
                                           'Babesia_microti_Data_Source',
                                           'Powassan_virus_Data_Source'], axis=1)
tick_path_df
```

```python
tick_path_df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
def change_string(status):
  if status == "Present":
    return 1
  else:
    return 0

le = LabelEncoder()
clean_tick_path_df = tick_path_df.copy()
clean_tick_path_df['Borrelia_burgdorferi_sensu_stricto_County_Status'] = clean_tick_path_df['Borrelia_burgdorferi_sensu_stricto_County_Status'].apply(change_string)
clean_tick_path_df['Borrelia_mayonii_County_Status'] = clean_tick_path_df['Borrelia_mayonii_County_Status'].apply(change_string)
clean_tick_path_df['Borrelia_miyamotoi_County_Status'] = clean_tick_path_df['Borrelia_miyamotoi_County_Status'].apply(change_string)
clean_tick_path_df['Anaplasma_phagocytophilum_County_Status'] = clean_tick_path_df['Anaplasma_phagocytophilum_County_Status'].apply(change_string)
clean_tick_path_df['Ehrlichia_muris_eauclairensis_County_Status'] = clean_tick_path_df['Ehrlichia_muris_eauclairensis_County_Status'].apply(change_string)
clean_tick_path_df['Babesia_microti_County_Status'] = clean_tick_path_df['Babesia_microti_County_Status'].apply(change_string)
clean_tick_path_df['Powassan_virus_County_Status'] = clean_tick_path_df['Powassan_virus_County_Status'].apply(change_string)
clean_tick_path_df['State'] = le.fit_transform(clean_tick_path_df['State'])
clean_tick_path_df
```
