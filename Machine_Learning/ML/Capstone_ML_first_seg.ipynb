```python
# Reading the 2021 records sheet into a csv file
tick_2021_records = pd.read_excel(io.BytesIO(data['Public_Use_Ixodes_County_Table_2022.xlsx']), 
                                  sheet_name='Ixodes records 2021', 
                                  header=1)
tick_2021_records

```

```python
# Checking the column names 
tick_2021_records.columns
```

```python
# Dropping columns that are not needed for analysis
cleaned_2021_records = tick_2021_records.drop(['County','Ixodes_scapularis_data_source','Ixodes_pacificus_data_source'], axis=1)
cleaned_2021_records
```

```python
cleaned_2021_records.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
def change_string(status):
  if status == "Reported":
    return 1
  elif status == "Established":
    return 2
  else:
    return 0

le = LabelEncoder()
cleaned_2021_records_2 = cleaned_2021_records.copy()
cleaned_2021_records_2['Ixodes_pacificus_county_status'] = cleaned_2021_records_2['Ixodes_pacificus_county_status'].apply(change_string)
cleaned_2021_records_2['Ixodes_scapularis_County_Status'] = cleaned_2021_records_2['Ixodes_scapularis_County_Status'].apply(change_string)
cleaned_2021_records_2['State'] = le.fit_transform(cleaned_2021_records_2['State'])
cleaned_2021_records_2
```
